# AI Study Pal Capstone: Feedback Model Training
This notebook demonstrates template-based feedback generation for quiz results.

```python
# 1. Feedback Generation Function
def generate_feedback(accuracy, subject):
    if accuracy >= 80:
        return f"Great job on {subject}! You're progressing well. Keep practicing medium difficulty questions."
    elif accuracy >= 60:
        return f"Good progress in {subject}, keep going! Focus on reviewing mistakes."
    else:
        return f"Don't give up on {subject}! Practice more easy questions and review concepts."

# Example usage
for acc in [90, 75, 50]:
    print(generate_feedback(acc, 'AIML Fundamentals'))
```

```python
# 2. Save Function for Backend Use
import joblib
joblib.dump(generate_feedback, 'feedback_func.pkl')
```
